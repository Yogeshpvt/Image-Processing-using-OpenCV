```python
from tkinter import filedialog 
import cv2
import sys
import tkinter
```

```python
def ProcessImage(self):
    
    Image = cv2.imread("leaf2.jpg")
    cv2.imshow("Diseased Image", Image)
    
    b = Image[:, :, 0] #extracting blue channel (0)
    g = Image[:, :, 1] #extracting green channel (1)
    r = Image[:, :, 2] #extracting red channel (2)
    
    cv2.imshow("Red Channel", r)
    cv2.imshow("Green Channel", g)
    cv2.imshow("Blue Channel", b)
    
    Disease = r - g #the pathological region will have higher red channel value compared to other 2 channels
    
    global Alpha
    Alpha = b
    GetAlpha(Image) #Function to get the alpha of the image
    
    cv2.imshow("Alpha Channel", Alpha)
    ProcessingValue = S.get()
    
    for i in range(0, Image.shape[0]):
        for j in range(0, Image.shape[1]):
            if int(g[i, j]) > ProcessingValue:
                Disease[i, j] = 255
                
    cv2.imshow("Highlighting the Diseased spots", Disease)
    
    DisplayDiseasePercentage(Disease) #Function to calculate diseased percentage
    
    S.bind('<ButtonRelease-1>', ProcessImage)
    MainWindow.mainloop()   
```

```python
def GetAlpha(Image):
    
    global Alpha
    for i in range(0, Image.shape[0]):
        for j in range(0, Image.shape[1]):
            
            if Image[i, j, 0] > 200 and Image[i, j, 1] > 200 and Image[i, j, 2] > 200:
                Alpha[i, j] = 255 #shades of white are made to white
            else:
                Alpha[i, j] = 0 #All the color values are made to black
```

```python
def DisplayDiseasePercentage(Disease):
   
    Count = 0
    Res = 0
    for i in range(0, Disease.shape[0]):
        for j in range(0, Disease.shape[1]):
            if Alpha[i, j] == 0:
                Res += 1
            if Disease[i, j] < S.get():
                Count += 1
    Percent = (Count / Res) * 100
    DiseasePercent.set("Percentage Disease: " + str(round(Percent, 2)) + "%")


Alpha = None
MainWindow = tkinter.Tk()
MainWindow.title("Plant Disease Detector")

#Tkinter scale
#The Scale widget is used to implement the graphical slider to the python application. 
#the user can slide through the range of values shown on the slider and select the one among them.

S = tkinter.Scale(MainWindow, from_=0, to=255, length=500, 
                  orient=tkinter.HORIZONTAL,
                  background='white', 
                  fg='black', 
                  troughcolor='white', 
                  label="Processing Value")

S.pack() #The pack() fill option is used to make a widget fill the entire frame
S.set(150)

DiseasePercent = tkinter.StringVar()
L = tkinter.Label(MainWindow, textvariable=DiseasePercent)
L.pack()

def GetFile():
    if len(sys.argv) > 1:
        return sys.argv[1]
    else:
        return filedialog.askopenfilename(title="Select Image")

filename = GetFile()
if filename != "":
    ProcessImage(None)
else:
    print("No File!")
    exit(0)    
```
